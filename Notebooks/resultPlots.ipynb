```python
import matplotlib.pyplot as plt
#import mplhep as hep
#plt.style.use(hep.cms.style.ROOT)
import ROOT
import PlotTools
from rootpy.plotting.style import get_style
from rootpy.plotting.style.cmstdr.labels import CMS_label
from ROOT import TLine, TFile, TCanvas, kRed, THStack, kYellow, kGreen, kBlack, TLegend, TGraphAsymmErrors
```

```python
file = ROOT.TFile.Open('../data/shapes_fit_Zprime_m500_gb1_dbs0.root','read')
file.cd()
```

```python
mjjsr = file.Get('all/Zp500')
#mjjsr_dcr = file.Get('mjjsr_dcr')
data = file.Get('all/data_obs')
backg = file.Get('all/backg')
```

```python
file350 = ROOT.TFile.Open('../data/shapes_fit_Zprime_m350_gb1_dbs0.root','read')
file350.cd()
mjjsr350 = file350.Get('all/Zp350')
file650 = ROOT.TFile.Open('../data/shapes_fit_Zprime_m650_gb1_dbs0.root','read')
file650.cd()
mjjsr650 = file650.Get('all/Zp650')
file800 = ROOT.TFile.Open('../data/shapes_fit_Zprime_m800_gb1_dbs0.root','read')
file800.cd()
mjjsr800 = file800.Get('all/Zp800')
file950 = ROOT.TFile.Open('../data/shapes_fit_Zprime_m950_gb1_dbs0.root','read')
file950.cd()
mjjsr950 = file950.Get('all/Zp950')
```

```python
style = get_style('CMSTDR')

with style:
    c1 = ROOT.TCanvas("","",700,700)
    c1.SetLogy()
    backg.SetStats(0)
    backg.SetTitle("")
    backg.GetYaxis().SetTitle('Events/10 GeV')
    backg.GetXaxis().SetTitle('m_{jj}[GeV]')
    backg.GetYaxis().SetLabelSize(0.02)
    backg.GetYaxis().SetTitleOffset(1.2)
    backg.GetXaxis().SetLabelSize(0.02)
    backg.GetXaxis().SetTitleOffset(1.2)
    backg.GetXaxis().SetTitleSize(0.03)
    backg.SetLineWidth(1)
    backg.SetFillColor(4)
    backg.SetLineColor(4)
    backg.SetMarkerColor(4)
    backg.SetFillColor(4)
    backg.SetFillStyle(3001)
    mjjsr.SetFillStyle(0)
    mjjsr350.SetFillStyle(0)
    mjjsr650.SetFillStyle(0)
    mjjsr800.SetFillStyle(0)
    mjjsr950.SetFillStyle(0)

    #backg.Rebin(2)
    #data.Rebin(2)
    #mjjsr.Rebin(2)
    #mjjsr350.Rebin(2)
    #mjjsr650.Rebin(2)
    #mjjsr800.Rebin(2)
    #mjjsr950.Rebin(2)
    mjjsr.SetLineColor(2)
    mjjsr.SetLineWidth(2)
    mjjsr350.SetLineColor(3)
    mjjsr350.SetLineWidth(2)
    #mjjsr350.Smooth()
    #mjjsr650.Smooth()
    #mjjsr800.Smooth()
    #mjjsr950.Smooth()
    mjjsr650.SetLineColor(4)
    mjjsr650.SetLineWidth(2)
    mjjsr800.SetLineColor(5)
    mjjsr800.SetLineWidth(2)
    mjjsr950.SetLineColor(6)
    mjjsr950.SetLineWidth(2)
    
    data.SetLineColor(kBlack)
    data.SetMarkerColor(kBlack)
    data.SetMarkerStyle(20)
    backg.GetYaxis().SetRangeUser(1,2000000)
    #backg.Draw("HISTE")
    
    #data.Draw('same')
    mjjsr350.Draw('HIST')
    mjjsr.Draw('HISTsame')
    mjjsr650.Draw('HISTsame')
    mjjsr800.Draw('HISTsame')
    mjjsr950.Draw('HISTsame')
    ROOT.gPad.Update()
    #graph.Draw()
    
    leg = TLegend(.59,.68,.81,.94)
    leg.SetBorderSize(0)
    leg.SetFillColor(0)
    leg.SetFillStyle(0)
    leg.SetTextFont(42)
    leg.SetTextSize(0.03)
    
    leg.AddEntry(mjjsr350,"Z' 350 GeV, #delta_{bs} = 0, g_{b} = 1","L")
    leg.AddEntry(mjjsr,"Z' 500 GeV, #delta_{bs} = 0, g_{b} = 1","L")
    leg.AddEntry(mjjsr650,"Z' 650 GeV, #delta_{bs} = 0, g_{b} = 1","L")
    leg.AddEntry(mjjsr800,"Z' 800 GeV, #delta_{bs} = 0, g_{b} = 1","L")
    leg.AddEntry(mjjsr950,"Z' 950 GeV, #delta_{bs} = 0, g_{b} = 1","L")
    #leg.AddEntry(data,"Data","APL")
    #leg.AddEntry(backg,"Background","f")
    leg.Draw("same")
    c1.SetGridx()
    c1.SetGridy()
    c1.Draw()
    CMS_label("", sqrts=13)
  
```
