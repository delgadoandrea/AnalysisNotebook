```python
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import numpy as np
import matplotlib.patches as mpatches
```

```python
di90 = [0.949,1.18,1.144,1.221,1.15]
qu45 = [0.574134,0.519629,0.843832,0.889085,0.81]
err = [0.0801839,0.038116455,0.0802222,0.0668643,0.0732]
m = [350,500,650,800,950]
ratio = np.divide(di90,qu45)
```

```python
star = mpath.Path.unit_regular_star(6)
circle = mpath.Path.unit_circle()
# concatenate the circle with an internal cutout of the star
verts = np.concatenate([circle.vertices, star.vertices[::-1, ...]])
codes = np.concatenate([circle.codes, star.codes])
cut_star = mpath.Path(verts, codes)

fig, (ax1, ax2) = plt.subplots(2,1)

ax1.plot(m,di90, '--r', marker=cut_star, markersize=15,label='Di90')
ax1.plot(m, qu45, '--b',marker=cut_star, markersize=15,label='quad45')
#plt.plot(m,ratio, '-', label='ratio')
#ax1.title('Signal acceptance (trigger + preselection + T4 tag)')
ax1.set_xlabel("m(Z')[GeV]")
ax1.set_ylabel('Acceptance(%)')
ax1.set_ylim(0.25,1.5)
ax1.legend()
ax1.grid(True)

ax2.plot(m,ratio)
ax2.set_xlabel("m(Z')[GeV]")
ax2.set_ylabel('Ratio')
ax2.grid(True)
ax2.set_ylim(0.75,3.0)
#plt.savefig('../plots/triggerAcc.png')
plt.show()
```

```python
kwargs = dict(capsize=2,
              elinewidth=1.1, linewidth=0.6, ms=7)
plt.errorbar(m, di90, yerr=err,fmt='ro',**kwargs, label = 'Di90')
plt.errorbar(m, qu45, yerr=err,fmt='bo',**kwargs, label = 'Quad45')

#plt.plot(m,di90, '--r', marker=cut_star, markersize=15,label='Di90')
#plt.plot(m, qu45, '--b',marker=cut_star, markersize=15,label='quad45')
#plt.plot(m,ratio, '-', label='ratio')
plt.title('Signal acceptance (trigger + preselection + T4 tag)')
plt.xlabel("m(Z')[GeV]")
plt.ylabel('Acceptance(%)')
#ax1.set_ylim(0.25,1.5)
plt.legend()
plt.grid(True)

plt.savefig('../plots/triggerAcc.png')
plt.show()
```
