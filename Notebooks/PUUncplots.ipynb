```python
deltabs = [0.0, 0.25, 0.5, 0.75, 1.0]
nom = [1.182284442,0.919821624,0.562336479,0.355370182,0.19377206,]
err = [0.009151236,0.007014799,0.004106174,0.002561773,0.001748039]
acc_up = [1.269597839,1.00594791,0.602993179,0.379653311,0.21240179]
acc_up_err = [0.009151236,0.007014799,0.004106174,0.002561773,0.001748039]
acc_down = [1.096314257,0.839958644,0.525437107,0.331256629,0.17590302]
acc_down_err = [0.009151236,0.007014799,0.004106174,0.002561773,0.001748039] 
```

```python
plt.figure()
kwargs = dict(capsize=2,
              elinewidth=1.1, linewidth=0.6, ms=7)
plt.errorbar(deltabs, nom, yerr=err,fmt='bo',**kwargs, label = 'Nominal')
plt.errorbar(deltabs,acc_up, yerr=acc_up_err,fmt='r^',**kwargs, label = 'Scale Up')
plt.errorbar(deltabs,acc_down, yerr=acc_down_err,fmt='gs',**kwargs, label = 'Scale Down')
plt.title('m = 500 GeV')
plt.xlabel('delta bs', fontsize=12)
plt.ylabel('Acceptance', fontsize=12)
plt.grid(True)
plt.legend(loc='best', frameon=True)
plt.savefig('../plots/PUUnc_deltabs.png')
plt.show()
```
